```python
import os

# Base directory matching your output
base_path = '/content/drive/MyDrive/Data Science/Last Assignment/Smart Survey Data Sets'

def scan_directory(directory_path):
    if not os.path.exists(directory_path):
        print(f"Error: The path '{directory_path}' does not exist.")
        return

    print(f"--- Scanning directory: {directory_path} ---\n")

    all_files = []
    as_files = []
    other_files = []
    directories_found = []

    for root, dirs, files in os.walk(directory_path):
        for d in dirs:
            # Track subdirectories
            rel_path = os.path.relpath(os.path.join(root, d), directory_path)
            directories_found.append(rel_path)
        for f in files:
            full_file_path = os.path.join(root, f)
            rel_file_path = os.path.relpath(full_file_path, directory_path)
            all_files.append(rel_file_path)
            if f.endswith('.as'):
                as_files.append(rel_file_path)
            else:
                other_files.append(rel_file_path)

    print(f"Total Folders Found: {len(directories_found)}")
    if directories_found:
        print("First 10 folders:")
        for folder in directories_found[:10]:
            print(f"  [Folder] {folder}")

    print(f"\nTotal Files Found: {len(all_files)}")
    print(f"Total '.as' (SMART survey) Files Found: {len(as_files)}")

    if as_files:
        print("\nFirst 15 '.as' files found:")
        for as_file in as_files[:15]:
            print(f"  - {as_file}")
    else:
        print("Warning: No '.as' files detected. Please verify if they are still compressed or have a different extension.")
        if other_files:
            print("\nShowing some other file types found instead:")
            for other in other_files[:10]:
                print(f"  - {other}")

# Run the scanner
scan_directory(base_path)
```

```python
import os
import pandas as pd
import numpy as np

file_path = '/content/drive/MyDrive/Data Science/Last Assignment/Smart Survey Data Sets/Final/admin2_surveys/individual_surveys/SOM_2022_11_DHOBLEY_admin2_individual.csv'
output_path = '/content/drive/MyDrive/Data Science/Last Assignment/Smart Survey Data Sets/Final/SOM_2022_11_DHOBLEY_household_aggregated.csv'

RECALL_PERIOD = 90  # Standard SMART recall period in days

def parse_wide_mortality(file_path):
    with open(file_path, 'r', encoding='utf-8', errors='ignore') as f:
        lines = f.readlines()

    mortality_lines = []
    found_headers = False
    headers = []

    for line in lines:
        cleaned = line.strip()
        if "P1_sex" in cleaned:
            headers = [col.strip() for col in cleaned.split(',')]
            found_headers = True
            continue

        if found_headers:
            # Break if we hit config settings or completely empty rows
            if not cleaned or cleaned.startswith("Training") or "Training" in cleaned:
                break
            parts = line.split(',')
            # Pad or trim parts to match header length precisely
            if len(parts) < len(headers):
                parts += [''] * (len(headers) - len(parts))
            else:
                parts = parts[:len(headers)]
            mortality_lines.append(parts)

    if not found_headers or not mortality_lines:
        return None

    df = pd.DataFrame(mortality_lines, columns=headers)
    return df

# Load the raw wide dataframe
df_raw = parse_wide_mortality(file_path)

if df_raw is not None and not df_raw.empty:
    print(f"Successfully loaded {len(df_raw)} household rows!")

    # Let's compute Household metrics
    aggregated_rows = []

    for idx, row in df_raw.iterrows():
        hh_id = row.get('HH', idx)
        cluster = row.get('Cluster', '')
        team = row.get('Team', '')

        total_hh_deaths = 0
        total_person_time = 0
        current_members_count = 0

        # Loop through up to 12 potential members
        for p in range(1, 13):
            prefix = f"P{p}_"
            sex_col = f"{prefix}sex"

            # Check if this member exists in the record
            if row.get(sex_col, '') != '':
                current_members_count += 1

                # Check status flags
                is_born = str(row.get(f"{prefix}born", '')).strip() == '1'
                is_died = str(row.get(f"{prefix}died", '')).strip() == '1'
                is_joined = str(row.get(f"{prefix}join", '')).strip() == '1'
                is_left = str(row.get(f"{prefix}left", '')).strip() == '1'

                # Track deaths
                if is_died:
                    total_hh_deaths += 1

                # Calculate Person-Time for this individual
                if is_born or is_died or is_joined or is_left:
                    total_person_time += (RECALL_PERIOD / 2)
                else:
                    total_person_time += RECALL_PERIOD

        aggregated_rows.append({
            'Cluster': cluster,
            'Team': team,
            'HH': hh_id,
            'Household_Size': current_members_count,
            'Total_Deaths': total_hh_deaths,
            'Total_Person_Time': total_person_time,
            # Target variable for ML (1 if any death occurred in the household, else 0)
            'Mortality_Target': 1 if total_hh_deaths > 0 else 0
        })

    df_agg = pd.DataFrame(aggregated_rows)
    df_agg.to_csv(output_path, index=False)
    print("\n--- Aggregation Complete ---")
    print(df_agg.head(5))
else:
    print("Failed to extract rows. Verify file headers or content manually.")
```

# **This section is the training section**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb
import pickle

# Load dataset
data_path = '/content/drive/MyDrive/Data Science/Last Assignment/Smart Survey Data Sets/Final/final_merged_analysis_dataset.csv'
df = pd.read_csv(data_path)

# Drop raw target-leaking fields like 'Total_Deaths'
# Total_Deaths tells the model the answer directly! We must drop it.
X_raw = df.drop(columns=['Total_Deaths', 'Mortality_Target', 'HH', 'Cluster'])
y = df['Mortality_Target']

# One-hot encode categorical features (District, Year, Month)
X_encoded = pd.get_dummies(X_raw, columns=['District', 'Year', 'Month'], drop_first=True)

# Split into Train and Test sets (Stratified to maintain 1.94% mortality ratio in both splits)
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42, stratify=y
)

# Standardize numerical features (Household_Size, Total_Person_Time)
scaler = StandardScaler()
num_cols = ['Household_Size', 'Total_Person_Time']
X_train[num_cols] = scaler.fit_transform(X_train[num_cols])
X_test[num_cols] = scaler.transform(X_test[num_cols])

# Compute scale factor for XGBoost class weighting
scale_pos_weight = (len(y_train) - sum(y_train)) / sum(y_train)

# Initialize the 3 selected models with balance weights
models = {
    'Logistic Regression': LogisticRegression(class_weight='balanced', max_iter=1000, random_state=42),
    'Random Forest': RandomForestClassifier(class_weight='balanced', n_estimators=100, random_state=42),
    'XGBoost': xgb.XGBClassifier(scale_pos_weight=scale_pos_weight, eval_metric='logloss', random_state=42)
}

# Train and print status
trained_models = {}
for name, model in models.items():
    print(f"Training {name}...")
    model.fit(X_train, y_train)
    trained_models[name] = model
    print(f"{name} trained successfully.")

# Save trained models and test data for Step 6 (Evaluation)
os.makedirs('/content/models', exist_ok=True)
with open('/content/models/trained_models.pkl', 'wb') as f:
    pickle.dump((trained_models, X_test, y_test), f)

print("\nAll models trained and saved to disk.")
```

```python
import os
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, recall_score, f1_score, roc_auc_score, confusion_matrix, roc_curve

# Load trained models and real split test data
with open('/content/models/trained_models.pkl', 'rb') as f:
    trained_models, X_test, y_test = pickle.load(f)

metrics_list = []

# Directory to save evaluation figures
os.makedirs('/content/figures', exist_ok=True)

plt.figure(figsize=(10, 8))

for name, model in trained_models.items():
    # Predict labels and probabilities
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]

    # Compute metrics
    acc = accuracy_score(y_test, y_pred)
    rec = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    auc = roc_auc_score(y_test, y_prob)

    cm = confusion_matrix(y_test, y_pred)

    metrics_list.append({
        'Model': name,
        'Accuracy': f"{acc:.2%}",
        'Recall': f"{rec:.2%}",
        'F1-Score': f"{f1:.2%}",
        'AUC': f"{auc:.4f}"
    })

    # Generate Confusion Matrix Heatmap
    plt.figure(figsize=(5, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
                xticklabels=['No Death', 'Death'], yticklabels=['No Death', 'Death'])
    plt.title(f'{name} Confusion Matrix')
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.tight_layout()
    plt.savefig(f'/content/figures/{name.replace(" ", "_").lower()}_cm.png')
    plt.close()

    # Track ROC Curve
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    plt.plot(fpr, tpr, label=f'{name} (AUC = {auc:.4f})')

# Save and show ROC Plot
plt.plot([0, 1], [0, 1], 'k--', label='Random Guess')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.tight_layout()
plt.savefig('/content/figures/roc_curve_comparison.png')
plt.show()

# Display Comparison Table
df_metrics = pd.DataFrame(metrics_list)
print("\n================ MODEL COMPARISON TABLE ================")
print(df_metrics.to_string(index=False))
print("========================================================")
```
